### Analizamos los outliers

Podemos transformar los datos usando escala logaritmica, normas. La ventaja es que podemos "normalizar" los datos y eliminar los outliers. No podemos aplicarlo para variables con valor menor o igual a 0.

```python
count_vals = list(df.columns)[2:-1]
```

```python
with_outliers = []
def outlier_count(col, data=df):
    q75, q25 = np.percentile(data[col], [75, 25])
    iqr = q75 - q25
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    if(outlier_count != 0):
        with_outliers.append(col)
        print(15*'-' + col + 15*'-')
        print('Cantidad de outliers: {}'.format(outlier_count))
        print('Porcentaje que son outliers: {}%'.format(outlier_percent))
    
for col in count_vals:
    outlier_count(col)
```

```python
plt.figure(figsize=(20,20))
for i, col in enumerate(count_vals, start=1):
    plt.subplot(6, 4, i)
    df.boxplot(col)
```

Observamos que 'Measles', 'GDP' y 'Population' tienen bastantes valores atípicos. De ellos,  sospechamos que 'Measles' (sarampión) y 'Population' (población) no serán muy útiles como estimadores: la sarampión no es una enfermedad letal en la vasta mayoría de los casos (ni perjudica mucho a las víctimas comparada con otras enfermedades más mortíferas), y la población sospechamos no afectará mucho la calidad de vida (luego estudiaremos cuán correlacionada está con esta métrica objetivo). 'GDP' (PBI, Producto Bruto Interno) se suele usar como una métrica para medir el 'progreso económico' de un país (aunque está lejos de ser infalible), de modo que esperamos que tenga cierta relación con el avance e instrucción tecnológica (equipamiento para hospitales y capacitación), que asumimos repercutirá sobre la calidad de vida (más tarde se estudiará cuán correlacionado está el GDP con la esperanza de vida).

Así, 'Measles' y 'Population' ya se perfilan como métricas a descartar.

#### Transformamos las variables con outliers usando logaritmo o normalizando restando la media y dividiendo por el desvío estándar

```python
with_outliers.remove("thinness 5-9 years")
with_outliers.remove("thinness 10-19 years")
with_outliers.remove("Measles")
for col in with_outliers:
    #Chequeamos que no hayan valores cero
    if((df[col] == 0).sum() == 0):
        df[col] = np.log(df[col])
    else:
        df[col] = (df[col]-np.mean(df[col]))/np.std(df[col])
```

Se puede ver que muchas de las variables tienen una correlación mediocre (~0.5) con la esperanza de vida, entre 
las cuales se encuentran muchas enfermedades y algunas métricas como consumo de alcohol. Si bien estas métricas no 
parecen irrelevantes (muchas de las enfermedades suelen afectar la calidad de vida, y el consumo de alcohol excesivo 
también puede repercutir negativamente en dicha métrica), es probable que individualmente no expliquen la calidad de 
vida, e incluirlas a todas para el modelado dispararía la cantidad de variables a considerar.

* 'Income composition of resources' tiene una muy buena correlación, pero no encontramos forma de completar sus valores faltantes usando la base de datos de la OMS (ncluso resultó difícil completar los valores con fuentes ajenas). Se descarta por falta de datos 'irrecuperables' y porque podremos usar otras métricas con correlaciones similarmente buenas.
* 'GDP' tiene muy buena correlación, a pesar de sus outliers. Dado que pudimos completarla e intuitivamente parecería relacionada a la calidad de vida, **será una de las variables a tener en cuenta para el modelado**.
* 'Schooling' muestra buena correlación, y parecería crítico incluir una métrica que mida la escolaridad (dada su repercusión sobre la cantidad de profesionales de la salud y tecnología disponible en un país). Dado que sus datos pudieron completarse con una fuente ajena a la OMS, **será una de las variables a tener en cuenta para el modelado**.
* 'Polio' muestra una correlación razonable, pero es una variable que tiene una correlación bastante fuerte con otras métricas como 'Schooling', 'Adult mortality' e 'Infant mortality', que sospechamos tendrán un mayor poder predictivo dado que se trata de métricas que afectan más aspectos de la vida de las personas, o revelan más acerca de un país.
* 'Diphtheria', ídem 'Polio'.
* 'Alcohol' tiene una correlación razonable, pero sospechamos que el consumo de alcohol no repercutirá demasiado sobre la calidad de vida de un país.
* 'Hepatitis B', 'Total expenditure', 'Population' y 'Measles' no parecen tener muy buenas correlaciones con la calidad de vida.
* 'thinness 5-9 years', 'thinness 10-19 years', 'Infant Mortality' 'Adult Mortality' y 'Under five Mortality' tienen correlaciones de buenas a excelentes con la calidad de vida, y resulta razonable que repercutan sobre la calidad de vida, por lo que **serán variables a tener en cuenta para el modelado**.
